# Space and Time

## Stability and the CFL condition

```python
import numpy as np
from matplotlib import pyplot as py
% matplotlib inline

# Font family and size to use for Matplotlib figures
py.rcParams['font.family'] = 'serif'
py.rcParams['font.size'] = 16
```

```python
'''
    Solves the 1D linear convection equation with constant
    speed c in the domain [0, L] and plots the solution
    along with the ICs
    
    nx : integer, number of grid points to discretize the domain
    L : float, optional, length of the domain, default: 2.0
    c : float, optional, convection speed, default: 1.0
    dt : float, optional, time-step size, default: 0.025
    nt : integer, optionl, number of time-steps to computer, default: 20
'''

def linear_conv(nx, L=2.0, c=1.0, dt=0.025, nt=20):
    # discretize the spatial grid
    dx = L/(nx-1)
    x = np.linspace(0.0, L, num = nx)
    
    # initial conditions
    u0 = np.ones(nx)
    mask = np.where(np.logical_and(x >= 0.5, x <= 1.0))
    u0[mask] = 2.0
    
    # integrate the solution in time
    u = u0.copy()
    for n in range(1, nt):
        u[1:] = u[1:] - c * dt/dx * (u[1:] - u[:-1])
    
    # plot of the solution along with the ICs
    py.figure(figsize = (5.0,5.0))
    py.xlabel('x')
    py.ylabel('y')
    py.grid()
    py.plot(x, u0, label = 'Initial Condition', color = 'C0', 
                linestyle = '--', linewidth = 2)
    py.plot(x, u, label = 'nt = {}'.format(nt), color = 'C1',
                linestyle = '-', linewidth = 2)
    py.legend()
    py.xlim(0.0, L)
    py.ylim(0.0, 2.5);
```

### Plots with increasingly fine mesh: 41, 61, 71, 85 points

```python
linear_conv(41) # 41 spatial grid points
```

```python
linear_conv(61) # 61 spatial grid points
```

```python
linear_conv(71) # 71 spatial grid points
```

```python
linear_conv(85) # 85 spatial grid points
```

Comment on 85 spatial grid points plot:
* Over the time period $\Delta t$, the wave is travelling a distrance which greater than `dx`, and the solution becomes *unstable*.
* The lenght `dx` of grid spacing is inversely proportional to the number of total points `nx`: more grid points, `dx` got smaller
* The numerical scheme is no longer possible to solve the equation once `dx` got smaller than the $c\Delta t$ - the distance travelled by the numerical solution in one time-step.

The mathematical domain of dependence of the solution of the original PDE is contained in the domain of dependence of the numerical scheme.

The following condition ensures the statement above.
$$ \begin{equation}
\sigma = \frac{c \Delta t}{\Delta x} \leq 1
\end{equation} $$

The stability of the numerical solution requires that step size `dt` is calculated with respect to the size of `dx` to satisfy the condition above. 

$c\Delta t/\Delta x$ is called the **Courant-Friedrichs-Lewy number** (CFL number), often denoted by $\sigma$.

The value $\sigma_{\text{max}}$ that will ensure stability depends on the discretization used; for the forward-time/backward-space scheme, the condition for stability is $\sigma<1$.

The code below uses the CFL nuber to calculate the appropriate time-step `dt` depending on the size of `dx`.

```python
'''
    Solves the 1D linear convection equation with constant
    speed c in the domain [0, L] and plots the solution
    along with the ICs.
    
    The time-step size is calculated
    based on a CFL contraint.
    
    nx : integer, number of grid points to discretize the domain
    L : float, optional, length of the domain, default: 2.0
    c : float, optional, convection speed, default: 1.0
    sig : float, optional, CFL constrain, default: 0.5
    dt : float, optional, time-step size, default: 0.025
    nt : integer, optionl, number of time-steps to computer, default: 20
'''

def lin_conv_cfl(nx, L = 2.0, c = 1.0, sig = 0.5, nt = 20):
    # discretize the spatial grid
    dx = L/(nx-1)
    x = np.linspace(0.0, L, num = nx)
    
    # time-step size based on the CFL constraint
    dt = sig * dx / c
    
    # initial conditions
    u0 = np.ones(nx)
    mask = np.where(np.logical_and(x >= 0.5, x <= 1.0))
    u0[mask] = 2.0
    
    # integrate the solution in time
    u = u0.copy()
    for n in range(1, nt):
        u[1:] = u[1:] - c * dt/dx * (u[1:] - u[:-1])
        
    # plot of the solution along with the ICs
    py.figure(figsize = (5.0,5.0))
    py.xlabel('x')
    py.ylabel('y')
    py.grid()
    py.plot(x, u0, label = 'Initial Condition', color = 'C0', 
                linestyle = '--', linewidth = 2)
    py.plot(x, u, label = 'nt = {}'.format(nt), color = 'C1',
                linestyle = '-', linewidth = 2)
    py.legend()
    py.xlim(0.0, L)
    py.ylim(0.0, 2.5);
```

It does not matter how many points are used for the spatial grid. The solution will always be stable.

```python
lin_conv_cfl(85)
```

```python
lin_conv_cfl(121)
```

As the number of `nx` increases, the wave convects a shorter and shorter distance.
