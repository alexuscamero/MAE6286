# Riding the Wave

## Conservation of mass
$\frac{D\,M}{Dt} =0$, closed system - > control volume

rate of change of mass in the control volume + rate of flow of mass across control surfaces $= 0$

Let mass desnity = $\rho$. Mass is equal to $\rho\times$ volume

1D velocity component by $u$, and conservation of mass for the control volume: 

$$
\begin{equation}
\frac{\partial}{\partial t}\int_{\text{cv}}\rho \, dV + \int_{\text{cs}}\rho \, u\, dA =0
\end{equation}
$$

    cv - control volume
    cs - control surface

rate of change of mass in cv + rate of flow of mass with velocity $u$ across cs = 0

for the second term: 
$$
\int_{\text{cs}}\rho \, u\, dA \rightarrow \frac{\partial}{\partial x}(\rho u) dA dx
$$

Thus, the 1D mass conservation equation in differential form:

$$
\begin{equation}
\frac{\partial \rho}{\partial t} + \frac{\partial}{\partial x}(\rho u)=0
\end{equation}
$$

## General Conservation Laws

$$
\vec{F}\cdot d\vec{A}
$$

- where $d\vec{A}$ points in the direction of the outward normal to the surface

The general conservation law for a quantity $e$ :

$$
\begin{equation}
\frac{\partial}{\partial t}\int_{\text{cv}}e \, dV + \oint_{\text{cs}}\vec{F}\cdot d\vec{A} =0
\end{equation}
$$

Non - linear convection equation: 

$$
\begin{equation}
\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} = 0
\end{equation}
$$

Conservation form of the non - linear convection equation: 

$$
\begin{equation}
\frac{\partial u}{\partial t} + \frac{\partial}{\partial x} \left(\frac{u^2}{2} \right) = 0
\end{equation}
$$

- with flux $F=\frac{u^2}{2}$

## Traffic Flow Model

- when road is empty, drivers will approach a maximum velocity $u_{max}$
- if road is completely jampacked, ($\rho \rightarrow \rho_{max}$), velocity goes to zero

Traffic density - transport equation: 

$$
\begin{equation}
\frac{\partial \rho}{\partial t} + \frac{\partial F}{\partial x} = 0
\end{equation}
$$

 - where $F$ is the *traffic flux*, which in the linear traffic-speed model is given by: 

$$
\begin{equation}
F = \rho u_{\rm max} \left(1-\frac{\rho}{\rho_{\rm max}}\right)
\end{equation}
$$

Initial Condition: 

$$
\begin{equation}
\rho(x,0) = \left\{
\begin{array}{cc}
\rho_{\rm max}\frac{x}{2} & 0 \leq x < 2  \\
0 & 2 \leq x \leq 4 \\
\end{array}
\right.
\end{equation}
$$

```python
import numpy as np
from matplotlib import pyplot as py
% matplotlib inline

# Set the font family and size to use for Matplotlib figures.
py.rcParams['font.family'] = 'serif'
py.rcParams['font.size'] = 16
```

```python
'''
    Computes the green light IC. It consists of a
    shock with a linear distribution behind it.
    
    x : np.ndarray, location on the road as a 1D array of floats
    rho_light : float, car density at stop light
    
    Returns:
    rho : np.ndarray, initial car density along the road
            as a 1D array of floats
'''

def rho_green(x, rho_light):
    rho = np.zeros_like(x)
    mask = np.where(x < 2.0)
    rho[mask] = rho_light * x[mask] / 2.0
    return rho
```

```python
# Parameters
nx = 81  # number of locations on the road
L = 4.0  # length of the road
dx = L / (nx - 1)  # distance between two consecutive locations
nt = 30  # number of time step to compute
u_max = 1.0  # maximum speed allowed on the road
rho_max = 10.0  # maximum car density allowed on the road
rho_light = 10.0  # car density at the stoplight

# Discretize the road
x = np.linspace(0.0, L, num = nx)

# Compute the initial traffic density
rho0 = rho_green(x, rho_light)
```

```python
# Plot the initial traffic car density on the road
py.figure(figsize = (6.0, 4.0))
py.xlabel(r'$x$')
py.ylabel(r'$\rho$')
py.grid()
py.plot(x, rho0, color='C0', linestyle='-', linewidth=2)
py.xlim(0.0, L)
py.ylim(-0.5, 11.0);
```

Cars should slowly start moving forward - the density profule should move to the right

```python
'''
    Computes the traffic flux, F = V * rho
    
    rho : np.ndarray, traffic density along the road
            as a 1D array of floats
    u_max : float, max speed allowed on the road
    rho_max : float, max car density allowed on the road
    
    Returns:
    F : np.ndarray, the traffic flux along the road
            as a 1D array of floats
'''

def flux(rho, u_max, rho_max):
    F = rho * u_max * (1.0 - rho / rho_max)
    return F
```

### Forward - time / Backward - space Scheme
$$
\begin{equation}
\frac{\rho^{n+1}_i- \rho^n_{i}}{\Delta t}+ \frac{F^n_{i}-F^n_{i-1}}{\Delta x}=0
\end{equation}
$$

```python
'''
    Computes the history of the traffic density on the
    road at a certain time given the initial traffic
    density
    
    rho0 : np.ndarray, the initial car density along
            the road as a 1D array of floats
    nt : integer, number of time steps to compute
    dt : float, time-step size to integrate
    dx : float, distance between two consecutive locations
    bc_value : float, constant density at the first station
    args : list or tuple, positional arguments to be passed
            to the flux function
            
    Returns: 
    rho_hist : list of np.ndarray objects, the history of
            the car density along the road
'''

def ftbs(rho9, nt, dt, dx, bc_value, *args):
    rho_hist = [rho0.copy()]
    rho = rho0.copy()
    for n in range(nt):
        # Compute the flux
        F = flux(rho, *args)
        # Advance in time
        rho[1:] = rho[1:] - dt/dx * (F[1:] - F[:-1])
        # Set the left BC
        rho[0] = bc_value
        # Record the time-step solution
        rho_hist.append(rho.copy())
    return rho_hist
```

```python
# Set the time-step size based on CFL limit
sigma = 1.0
dt = sigma * dx / u_max  # time-step size

# compute the traffic density at all time-steps
rho_hist = ftbs(rho0, nt, dt, dx, rho0[0], u_max, rho_max)
```

```python
from matplotlib import animation
from IPython.display import HTML
```

```python
# Plot the initial traffic density
fig = py.figure(figsize=(6.0, 4.0))
py.xlabel(r'$x$')
py.ylabel(r'$\rho$')
py.grid()
line = py.plot(x, rho0, color='C0', linestyle='-', linewidth=2)[0]
py.xlim(0.0, L)
py.ylim(-0.5, 11.0)
fig.tight_layout()
```

```python
'''
    Update the line y-data of the Matplotlib figure
    
    n : integer, time-step index
    rho_hist : list of np.ndarray objects,
            history of the numerical solution
'''

def update_plot(n, rho_hist):
    fig.suptitle('Time Step {:0>2}'.format(n))
    line.set_ydata(rho_hist[n])
```

```python
# Create an animation of the traffic density
anim = animation.FuncAnimation(fig, update_plot, frames = nt,
                        fargs = (rho_hist,), interval = 100)

# Display the video
HTML(anim.to_html5_video())
```

From the equation, 
$$
\begin{equation}
\frac{\partial \rho}{\partial t} + \frac{\partial F}{\partial x} = 0
\end{equation}
$$

Using the chain rule,
$$
\begin{equation}
\frac{\partial \rho}{\partial t} + \frac{\partial F}{\partial \rho} \frac{\partial \rho}{\partial x} = 0
\end{equation}
$$

This is a wave of equation where the wave speed is $u_{\rm wave} = \frac{\partial F}{\partial\rho}$

That terms is:

$$
\begin{equation}
u_{\rm wave} = \frac{\partial F}{\partial \rho} = u_{\rm max} \left( 1-2\frac{\rho}{\rho_{\rm max}} \right)
\end{equation}
$$

The wave speed changes sign at $\rho = \rho_{\rm max}/2$. Thus, for the initial condition given for the green - light problem, the part of the wave under $\rho = \rho_{\rm max}/2$ will want to move right, whereas the part of the wave over this mark, will move left.

A scheme that is backward in space is $unstable$ for negative values of the wave speed.

## Upwind Schemes

Forward - time / backward- space scheme with the initial conditions: 

\begin{equation}\rho(x,0) = \left\{ \begin{array}{cc}
2.5 x & 0 \leq x < 2  \\
0 & 2 \leq x \leq 4 \\ \end{array} \right.\end{equation}

If all values of $\rho \leq \rho_{\rm max}/2$, then $\frac{\partial F}{\partial \rho}$ is positive everywhere. Thus, all wave speeds are positive.

```python
# Modified Parameters
nt = 40  # number of time step to compute
rho_light = 5.0  # car density at the stoplight

# Compute the initial traffic density
rho0 = rho_green(x, rho_light)

# Plot the initial traffic density
fig = py.figure(figsize=(6.0, 4.0))
py.xlabel(r'$x$')
py.ylabel(r'$\rho$')
py.grid()
line = py.plot(x, rho0,
                   color='C0', linestyle='-', linewidth=2)[0]
py.hlines(rho_max / 2.0, 0.0, L,
              label=r'$\rho_{max} / 2$',
              color='black', linestyle='--', linewidth=2)
py.legend()
py.xlim(0.0, L)
py.ylim(-0.5, 11.0)
fig.tight_layout()
```

```python
# Compute the traffic density at all time steps
rho_hist = ftbs(rho0, nt, dt, dx, rho0[0], u_max, rho_max)
```

```python
# Create an animation of the traffic density
anim = animation.FuncAnimation(fig, update_plot, frames = nt,
                               fargs = (rho_hist,), interval = 100)

# Display the animation
HTML(anim.to_html5_video())
```

A forward - time / forward - space scheme results a stable for negative wave speeds but unstable for positve wave speeds. So, make sure the wave speed is negative everywhere, ie. $u_{\rm wave} = \frac{\partial F}{\partial \rho} < 0 \ \forall \ x$

## Beware the CFL!

For a constant wave speed $u_{\rm wave} = c$:

$$
\begin{equation}
\sigma = c\frac{\Delta t}{\Delta x} < 1
\end{equation}
$$

The wave speed is space- and time-dependent, $u_{\rm wave} = u_{\rm wave}(x,t)$, and the CFL condition needs to apply for every point in space, at every instant of time. We just need $\sigma>1$ in one spot, for the whole solution to blow up.

Generalizing the CFL condition to: 
$$
\begin{equation}
\sigma = \max\left[ \left| u_{\rm wave} \right| \frac{\Delta t}{\Delta x} \right] < 1
\end{equation}
$$

which in our case is

$$
\begin{equation}
\sigma = \max\left[ u_{\rm max} \left| 1-\frac{2 \rho}{\rho_{\rm max}} \right| \frac{\Delta t}{\Delta x} \right] < 1
\end{equation}
$$

Thus, the closer $\rho$ is to zero, the more likely it is to be unstable.

```python
# Parameters
rho_light = 4.0
u_max = 1.1

# Compute the initial traffic density.
rho0 = rho_green(x, rho_light)

# Compute the traffic density at all time steps.
rho_hist = ftbs(rho0, nt, dt, dx, rho0[0], u_max, rho_max)

# Create an animation of the traffic density.
anim = animation.FuncAnimation(fig, update_plot, frames = nt,
                               fargs = (rho_hist,), interval = 100)

# Display the video.
HTML(anim.to_html5_video())
```

Failed. Only by changing $u_{\rm max}$ to $1.1$, even an algorithm that we know is stable for this problem, fails.

Since we kept $\Delta t/\Delta x=1$, the CFL number for $\rho=0$ is $1.1$. See where the instability begins?
