# Riding the Wave

## Numerical Schemes for Hyperbolic PDEs

Forward-time / backward-space difference scheme: cannot deal with wave speeds that move in more than one direction

It is also first - order accurate in space and time, not good enough.

## Red Light!

For a moving shock wave...

Imagine a very busy road and a red light at x = 4. Cars accumulate quickly in the front, where we have the maximum allowed density of cars between x = 3 and x = 4, and there is an incoming traffic of 50% the maximum allowed density $(\rho = 0.5\rho_{\rm max})$. 

Initial Condition:

$$
\begin{equation}
\rho(x,0) = \left\{
\begin{array}{cc}
0.5 \rho_{\rm max} & 0 \leq x < 3  \\
\rho_{\rm max} & 3 \leq x \leq 4 \\
\end{array}
\right.
\end{equation}
$$

```python
import numpy as np
from matplotlib import pyplot as py
% matplotlib inline

# Font family and size to use for Matplotlib figures
py.rcParams['font.family'] = 'serif'
py.rcParams['font.size'] = 16
```

```python
'''
    Computes the "red light" initial condition with shock
    
    Parameters:
    x : np.ndarray, location on the road as a 1D array of floats
    rho_max : float, maximum traffic density allowed
    
    Returns:
    rho : np.ndarray, the initial car density along road
        as a 1D array of floats
'''

def rho_red(x, rho_max):
    rho = rho_max * np.ones_like(x)
    mask = np.where( x < 3.0)
    rho[mask] = 0.5 * rho_max
    return rho
```

```python
# Parameters
# Set parameters.
nx = 81   # number of locations on the road
L = 4.0   # length of the road
dx = L / (nx - 1)   # distance between two consecutive locations
nt = 40   # number of time steps to compute
rho_max = 10.0   # maximum taffic density allowed
u_max = 1.0   # maximum speed traffic

# Get the road locations.
x = np.linspace(0.0, L, num=nx)

# Compute the initial traffic density.
rho0 = rho_red(x, rho_max)
```

```python
# Plot of the initial traffic density
fig = py.figure(figsize = (6.0, 4.0))
py.xlabel(r'$x$')
py.ylabel(r'$\rho$')
py.grid()
line = py.plot(x, rho0,
                   color='C0', linestyle='-', linewidth=2)[0]
py.xlim(0.0, L)
py.ylim(4.0, 11.0)
py.tight_layout()
```

Before we do any coding, let's think about the equation a little bit. The wave speed $u_{\rm wave}$ is $-1$ for $\rho = \rho_{\rm max}$ and $\rho \leq \rho_{\rm max}/2$, making all velocities negative. We should see a solution moving left, maintaining the shock geometry.

```python
'''
    Computes the traffic flux, F = V * rho
    
    Parameters:
    rho : np.ndarray
    u_max : float, max speed allowed on the road
    rho_max : float, max car density allowed on the road
    
    Returns:
    F : np.ndarray, traffic flux along the road as
            a 1D array of floats
'''

def flux(rho, u_max, rho_max):
    F = rho * u_max * (1.0 - rho / rho_max)
    return F
```

```python
from matplotlib import animation
from IPython.display import HTML
```

```python
'''
    Updates the line y-data of the Matplotlib figure
    
    Parameters:
    n : integer, time-step index
    rho_hist : list of np.ndarray objects, history of the
                    numerical solution
'''

def update_plot(n, rho_hist):
    fig.suptitle('Time Step {:0>2}'.format(n))
    line.set_ydata(rho_hist[n])
```

## Lax - Friedrichs Scheme

Conservation law for vehicle traffic, resulting in the following equation for the traffic density: 

$$
\begin{equation}
\frac{\partial \rho}{\partial t} + \frac{\partial F}{\partial x} = 0
\end{equation}
$$

$F$ is the *traffic flux*, which in the linear traffic-speed model is given by: 

$$
\begin{equation}
F = \rho u_{\rm max} \left(1-\frac{\rho}{\rho_{\rm max}}\right)
\end{equation}
$$

In the time variable, the natural choice for discretization is always a forward-difference formula; time invariably moves forward!

$$
\begin{equation}
\frac{\partial \rho}{\partial t}\approx \frac{1}{\Delta t}( \rho_i^{n+1}-\rho_i^n )
\end{equation}
$$

As is usual, the discrete locations on the 1D spatial grid are denoted by indices $i$ and the discrete time instants are denoted by indices $n$.

In a convection problem, using first-order discretization in space leads to excessive numerical diffusion.

The simplest approach to get second - accuraxy in space is to use central difference: 

$$
\begin{equation}
\frac{\partial F}{\partial x} \approx \frac{1}{2\Delta x}( F_{i+1}-F_{i-1})
\end{equation}
$$

But combining these two choices for time and space discretization in the convection equation has catastrophic results! The "forward-time, central scheme" (FTCS) is **unstable**.

**Lax - Friedrichs scheme** - replaces the solution value at $\rho^n_i$ by the average of the values at the neighboring grid points. If we do that replacement, we get the following discretized equation: 

$$
\begin{equation}
\frac{\rho_i^{n+1}-\frac{1}{2}(\rho^n_{i+1}+\rho^n_{i-1})}{\Delta t} = -\frac{F^n_{i+1}-F^n_{i-1}}{2 \Delta x}
\end{equation}
$$

The difference formula no longer uses the value at $\rho^n_i$ to obtain $\rho^{n+1}_i$. The stencil of the Lax-Friedrichs scheme is slightly different than that for the forward-time, central scheme.

This numerical discretization is **stable**. Unfortunately, substituting  $\rho^n_i$ by the average of its neighbors introduces a first-order error.

To implement the scheme in code, we need to isolate the value at the next time step, $\rho^{n+1}_i$, so we can write a time-stepping loop:

$$
\begin{equation}
\rho_i^{n+1} = \frac{1}{2}(\rho^n_{i+1}+\rho^n_{i-1}) - \frac{\Delta t}{2 \Delta x}(F^n_{i+1}-F^n_{i-1})
\end{equation}
$$

The function below implements Lax-Friedrichs for our traffic model.

To display animations, we're going to hold the results of each time step in the variable `rho`, a 2D array.  The resulting array `rho_n` has `nt` rows and `nx` columns.

```python
'''
    Computes the traffic density on the road at a certain
        time given the initial traffic density.
        Integration using Lax-Friedrichs scheme.
        
    Parameters:
    rho0 : np.ndarray, the initial traffic density along the
            road as a 1D array of floats
    nt : integer, number of time steps to compute
    dt : float, time-step size to integrate
    dx : float, distane between two consecutive locations
    bc_values : 2-tuple of floats, value of the density at
            the first and last locations
    args : list or tuple, positional arguments to be passed
            to the flux function
            
    Returns:
    rho_hist : list of np.ndarray objects, history of the
            car density along the road
'''

def friedrichs(rho0, nt, dt, dx, bc_values, *args):
    rho_hist = [rho0.copy()]
    rho = rho0.copy()
    for n in range(nt):
        # Compute the flux
        F = flux(rho, *args)
        # Advance in time using Lax-Friedrichs scheme
        rho[1:-1] = (0.5 * (rho[:-2] + rho[2:]) -
                     dt / (2.0 * dx) * (F[2:] - F[:-2]))
        # Set the value at the first location.
        rho[0] = bc_values[0]
        # Set the value at the last location.
        rho[-1] = bc_values[1]
        # Record the time-step solution.
        rho_hist.append(rho.copy())
    return rho_hist
```

### **Lax - Friedrichs** with $\frac{\Delta t}{\Delta x} = 1$

with CFL = 1

```python
# Set the time-step size based on CFL limit
sigma = 1.0
dt = sigma * dx / u_max   # time-step size

# Compute the traffic density at all time steps.
rho_hist = friedrichs(rho0, nt, dt, dx, (rho0[0], rho0[-1]), u_max, rho_max)

# Create animation of the traffic density
anim = animation.FuncAnimation(fig, update_plot, frames = nt,
                        fargs = (rho_hist,), interval = 100)
# Display the video.
HTML(anim.to_html5_video())
```

### **Lax - Friedrichs** with $\frac{\Delta t}{\Delta x} = 0.5$

```python
# Set the time-step size based on CFL limit
sigma = 0.5
dt = sigma * dx / u_max   # time-step size

# Compute the traffic density at all time steps.
rho_hist = friedrichs(rho0, nt, dt, dx, (rho0[0], rho0[-1]), u_max, rho_max)

# Create animation of the traffic density
anim = animation.FuncAnimation(fig, update_plot, frames = nt,
                        fargs = (rho_hist,), interval = 100)
# Display the video.
HTML(anim.to_html5_video())
```

The **_staircase_** behavior is sometimes called "odd-even decoupling."

## Lax - Wendroff Scheme

In recap, the Lax-Friedrichs method stabilizes the central difference in space for convection, but loses an order of accuracy.

First-order methods are just not good enough for convection problems, especially when you have sharp gradients (shocks).

The **Lax - Wendroff Scheme** achieves second-order accuracy in both space and time.

Starting with the Taylor series expansion (in the time variable) about $\rho^{n+1}$:

$$
\begin{equation}
\rho^{n+1} = \rho^n + \frac{\partial\rho^n}{\partial t} \Delta t + \frac{(\Delta t)^2}{2}\frac{\partial^2\rho^n}{\partial t^2} + \ldots
\end{equation}
$$

For the conservation law with $F=F(\rho)$, and using chain rule,

$$
\begin{equation}
\frac{\partial \rho}{\partial t} = -\frac{\partial F}{\partial x} = -\frac{\partial F}{\partial \rho} \frac{\partial \rho}{\partial x} = -J \frac{\partial \rho}{\partial x}
\end{equation}
$$

where 

$$
\begin{equation}
J = \frac{\partial F}{\partial \rho} = u _{\rm max} \left(1-2\frac{\rho}{\rho_{\rm max}} \right)
\end{equation}
$$

is the _Jacobian_ for the traffic model.

Next,

$$
\begin{equation}
\frac{\partial F}{\partial t} = \frac{\partial F}{\partial \rho} \frac{\partial \rho}{\partial t} = J \frac{\partial \rho}{\partial t} = -J \frac{\partial F}{\partial x}
\end{equation}
$$

In the last step above, the differential equation of the traffic model was used to replace the time derivative by a spatial derivative. These equivalences imply that

$$
\begin{equation}
\frac{\partial^2\rho}{\partial t^2} = \frac{\partial}{\partial x} \left( J \frac{\partial F}{\partial x} \right)
\end{equation}
$$

So, in the Taylor expansion: 

$$
\begin{equation}
\rho^{n+1} = \rho^n - \frac{\partial F^n}{\partial x} \Delta t + \frac{(\Delta t)^2}{2} \frac{\partial}{\partial x} \left(J\frac{\partial F^n}{\partial x} \right)+ \ldots
\end{equation}
$$

Reorganizing and discretizing the spatial derivatives with central differences, we get:

$$
\begin{equation}
\frac{\rho_i^{n+1} - \rho_i^n}{\Delta t} = -\frac{F^n_{i+1}-F^n_{i-1}}{2 \Delta x} + \frac{\Delta t}{2} \left(\frac{(J \frac{\partial F}{\partial x})^n_{i+\frac{1}{2}}-(J \frac{\partial F}{\partial x})^n_{i-\frac{1}{2}}}{\Delta x}\right)
\end{equation}
$$

Approximating the rightmost term inside the parenthesis, 

$$ \begin{equation}
\frac{J^n_{i+\frac{1}{2}}\left(\frac{F^n_{i+1}-F^n_{i}}{\Delta x}\right)-J^n_{i-\frac{1}{2}}\left(\frac{F^n_i-F^n_{i-1}}{\Delta x}\right)}{\Delta x}
\end{equation} $$

Then evaluate the Jacobian at the midpoints by using averages of the points on either side:

$$ \begin{equation}\frac{\frac{1}{2 \Delta x}(J^n_{i+1}+J^n_i)(F^n_{i+1}-F^n_i)-\frac{1}{2 \Delta x}(J^n_i+J^n_{i-1})(F^n_i-F^n_{i-1})}{\Delta x}.
\end{equation} $$

So, the equation becomes: 

$$\begin{align}
&\frac{\rho_i^{n+1} - \rho_i^n}{\Delta t} =  
-\frac{F^n_{i+1}-F^n_{i-1}}{2 \Delta x} + \cdots \\ \nonumber 
&+ \frac{\Delta t}{4 \Delta x^2} \left( (J^n_{i+1}+J^n_i)(F^n_{i+1}-F^n_i)-(J^n_i+J^n_{i-1})(F^n_i-F^n_{i-1})\right)
\end{align} $$

Solving for $\rho_i^{n+1}$:

$$ \begin{align}
&\rho_i^{n+1} = \rho_i^n - \frac{\Delta t}{2 \Delta x} \left(F^n_{i+1}-F^n_{i-1}\right) + \cdots \\ \nonumber 
&+ \frac{(\Delta t)^2}{4(\Delta x)^2} \left[ (J^n_{i+1}+J^n_i)(F^n_{i+1}-F^n_i)-(J^n_i+J^n_{i-1})(F^n_i-F^n_{i-1})\right]
\end{align} $$

with

$$ \begin{equation}J^n_i = \frac{\partial F}{\partial \rho} = u_{\rm max} \left(1-2\frac{\rho^n_i}{\rho_{\rm max}} \right).
\end{equation} $$

```python
'''
    Computes the Jacobian for the traffic model.
    
    Parameters:
    rho : np.ndarray, traffic density along the road as
            a 1D array of floats
    u_max : float, max speed allowed on the road
    rho_max : float, max car density allowed on the road
    
    Returns:
    J : np.ndarray, the Jacobian as a 1D array of floats
'''

def Jacobian(rho, u_max, rho_max):
    J = u_max * (1.0 - 2.0 * rho / rho_max)
    return J
```

```python
'''
    Computes the traffic density on the road at a certain
        time given the initial traffic density.
        Integration using Lax - Wendroff scheme.
        
    Parameters:
    rho0 : np.ndarray, the initial traffic density along the
            road as a 1D array of floats
    nt : integer, number of time steps to compute
    dt : float, time-step size to integrate
    dx : float, distane between two consecutive locations
    bc_values : 2-tuple of floats, value of the density at
            the first and last locations
    args : list or tuple, positional arguments to be passed
            to the flux function
            
    Returns:
    rho_hist : list of np.ndarray objects, history of the
            car density along the road
'''

def wendroff(rho0, nt, dt, dx, bc_values, *args):
    rho_hist = [rho0.copy()]
    rho = rho0.copy()
    for n in range(nt):
        # Compute the flux
        F = flux(rho, *args)
        # Compute the Jacobian
        J = Jacobian(rho, *args)
        # Advance in time using Lax-Wendroff scheme
        rho[1:-1] = (rho[1:-1] - dt / (2.0 * dx)
                     * (F[2:] - F[:-2]) + dt**2 / (4.0 * dx**2)
                     * ((J[1:-1] + J[2:]) * (F[2:] - F[1:-1])
                        - (J[:-2] + J[1:-1]) * (F[1:-1] - F[:-2])))
        # Set the value at the first location.
        rho[0] = bc_values[0]
        # Set the value at the last location.
        rho[-1] = bc_values[1]
        # Record the time-step solution.
        rho_hist.append(rho.copy())
    return rho_hist
```

### **Lax - Wendroff** with $\frac{\Delta t}{\Delta x} = 1$

```python
# Set the time-step size based on CFL limit
sigma = 1.0
dt = sigma * dx / u_max   # time-step size

# Compute the traffic density at all time steps.
rho_hist = wendroff(rho0, nt, dt, dx, (rho0[0], rho0[-1]), u_max, rho_max)

# Create animation of the traffic density
anim = animation.FuncAnimation(fig, update_plot, frames = nt,
                        fargs = (rho_hist,), interval = 100)
# Display the video.
HTML(anim.to_html5_video())
```

The **Lax - Wendroff** method captures the sharpness of the shock much better than Lax - Friedrichs scheme.

However, a strange wiggle appears right at the tail of the shock.

This is typical of many second-order methods as they introduce numerical oscillations where the solution is not smooth.

### **Lax - Wendroff** with $\frac{\Delta t}{\Delta x} = 0.5$

```python
# Set the time-step size based on CFL limit
sigma = 0.5
dt = sigma * dx / u_max   # time-step size

# Compute the traffic density at all time steps.
rho_hist = wendroff(rho0, nt, dt, dx, (rho0[0], rho0[-1]), u_max, rho_max)

# Create animation of the traffic density
anim = animation.FuncAnimation(fig, update_plot, frames = nt,
                        fargs = (rho_hist,), interval = 100)
# Display the video.
HTML(anim.to_html5_video())
```

The numerical oscillations got worse.

## MacCormack Scheme

The main drawback of the Lax-Wendroff method is having to calculate the Jacobian in every time step. With more complicated equations (like the Euler equations), calculating the Jacobian is a large computational expense.

The MacCormack scheme is a two-step method, in which the first step is called a _predictor_ and the second step is called a _corrector_. It achieves second-order accuracy in both space and time. One version is as follows:  

$$
\begin{equation}
\rho^*_i = \rho^n_i - \frac{\Delta t}{\Delta x} (F^n_{i+1}-F^n_{i}) \ \ \ \ \ \ \text{(predictor)}
\end{equation}
$$

$$
\begin{equation}
\rho^{n+1}_i = \frac{1}{2} (\rho^n_i + \rho^*_i - \frac{\Delta t}{\Delta x} (F^*_i - F^{*}_{i-1})) \ \ \ \ \ \ \text{(corrector)}
\end{equation}
$$

It appears like the first step is a forward-time/forward-space scheme, and the second step is like a forward-time/backward-space scheme (these can also be reversed), averaged with the first result.

You can compute problems with left-running waves and right-running waves, and the MacCormack scheme gives you a stable method (subject to the CFL condition).

```python
'''
    Computes the traffic density along the road as a 
        1D array of floats.
    
    Parameters:
    rho0 : np.ndarray, the initial traffic density along the
            road as a 1D array of floats
    nt : integer, number of time steps to compute
    dt : float, time-step size to integrate
    dx : float, distane between two consecutive locations
    bc_values : 2-tuple of floats, value of the density at
            the first and last locations
    args : list or tuple, positional arguments to be passed
            to the flux function
            
    Returns:
    rho_hist : list of np.ndarray objects, history of the
            car density along the road
'''

def maccormack(rho0, nt, dt, dx, bc_values, *args):
    rho_hist = [rho0.copy()]
    rho = rho0.copy()
    rho_star = rho.copy()
    for n in range(nt):
        # Compute the flux
        F = flux(rho, *args)
        # Predictor step of the MacCormack scheme
        rho_star[1:-1] = (rho[1:-1] - dt / dx * (F[2:] - F[1:-1]))
        # Compute the flux
        F = flux(rho_star, *args)
        # Corrector step of the MacCormack scheme
        rho[1:-1] = 0.5 * (rho[1:-1] + rho_star[1:-1] - dt / dx * (F[1:-1] - F[:-2]))
        # Set the value at the first location.
        rho[0] = bc_values[0]
        # Set the value at the last location.
        rho[-1] = bc_values[1]
        # Record the time-step solution.
        rho_hist.append(rho.copy())
    return rho_hist
```

### **MacCormack** with $\frac{\Delta t}{\Delta x} = 1$

```python
# Set the time-step size based on CFL limit
sigma = 1.0
dt = sigma * dx / u_max  # time-step size

# Compute the traffic density at all time steps
rho_hist = maccormack(rho0, nt, dt, dx, (rho0[0],
                            rho0[-1]), u_max, rho_max)

# Create an animation of the traffic density
anim = animation.FuncAnimation(fig, update_plot, frames = nt,
                               fargs = (rho_hist,), interval = 100)
# Display the video
HTML(anim.to_html5_video())
```

### **MacCormack** with $\frac{\Delta t}{\Delta x} = 0.5$

```python
# Set the time-step size based on CFL limit
sigma = 0.5
dt = sigma * dx / u_max  # time-step size

# Compute the traffic density at all time steps
rho_hist = maccormack(rho0, nt, dt, dx, (rho0[0],
                            rho0[-1]), u_max, rho_max)

# Create an animation of the traffic density
anim = animation.FuncAnimation(fig, update_plot, frames = nt,
                               fargs = (rho_hist,), interval = 100)
# Display the video
HTML(anim.to_html5_video())
```

For schocks, the best resolution will occur when the difference in the predictor step is in the direction of propagation.
